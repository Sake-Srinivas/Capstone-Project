```python
from IPython.display import Audio, display, Javascript
import io
import base64
from google.colab import output
import ffmpeg
import subprocess

# JavaScript for recording with a stop button
def record_audio_js():
    RECORD_BUTTON = """
    const sleep = time => new Promise(resolve => setTimeout(resolve, time))
    const b2text = blob => new Promise(resolve => {
      const reader = new FileReader()
      reader.onloadend = e => resolve(e.target.result)
      reader.readAsDataURL(blob)
    })

    var record = () => {
      return new Promise(async resolve => {
        let stream = await navigator.mediaDevices.getUserMedia({audio: true});
        let recorder = new MediaRecorder(stream);
        let chunks = [];

        recorder.ondataavailable = e => chunks.push(e.data);

        recorder.start();
        console.log("Recording started... Click Stop button to stop.");

        const stop_button = document.createElement("button");
        stop_button.textContent = "Stop Recording";
        document.body.appendChild(stop_button);

        stop_button.onclick = () => {
          recorder.stop();
          stop_button.remove();
        };

        recorder.onstop = async () => {
          let blob = new Blob(chunks);
          let text = await b2text(blob);
          resolve(text);
        };
      });
    }
    """
    display(Javascript(RECORD_BUTTON))

# Python function to call the recording JavaScript
def start_recording():
    record_audio_js()
    audio_string = output.eval_js('record()')
    audio_data = io.BytesIO(base64.b64decode(audio_string.split(',')[1]))
    return audio_data

# Function to save the recorded audio directly as MP3 format using ffmpeg
def save_audio(file_name="output"):
    audio_data = start_recording()

    # Write the audio data to a temporary file
    with open(f"{file_name}.webm", 'wb') as webm_file:
        webm_file.write(audio_data.read())

    # Use ffmpeg to convert from webm (or whatever format browser uses) to mp3
    try:
        subprocess.run([
            'ffmpeg', '-i', f'{file_name}.webm', f'{file_name}.mp3'
        ], check=True)
        print(f"Recording saved as {file_name}.mp3")
    except subprocess.CalledProcessError as e:
        print(f"An error occurred: {e}")

# Start recording and stop with the "Stop Recording" button
save_audio("my_recording")
```

```python
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
from reportlab.lib import colors
from reportlab.lib.units import inch

def generate_hospital_report(patient_name, age, gender, report_text, file_name="patient_report.pdf"):
    # Create a PDF with A4 size
    c = canvas.Canvas(file_name, pagesize=A4)
    width, height = A4

    # Header: Hospital Name
    c.setFont("Helvetica-Bold", 24)
    c.drawCentredString(width / 2.0, height - 100, "Shovin Hospitals")

    # Draw a line below the hospital name
    c.setStrokeColor(colors.black)
    c.setLineWidth(2)
    c.line(50, height - 110, width - 50, height - 110)

    # Subheader: Patient Details
    c.setFont("Helvetica", 14)

    # Name label and value
    c.drawString(50, height - 150, "Name: ")
    c.drawString(150, height - 150, patient_name)

    # Age label and value
    c.drawString(50, height - 180, "Age: ")
    c.drawString(150, height - 180, str(age))

    # Gender label and value
    c.drawString(50, height - 210, "Gender: ")
    c.drawString(150, height - 210, gender)

    # Add a line below the patient details section
    c.setLineWidth(1)
    c.line(50, height - 220, width - 50, height - 220)

    # Report Section
    c.setFont("Helvetica-Bold", 16)
    c.drawString(50, height - 250, "Patient Report:")

    # Patient report content
    c.setFont("Helvetica", 12)
    text_object = c.beginText(50, height - 280)
    text_object.setTextOrigin(50, height - 280)
    text_object.setLeading(14)  # Line spacing
    text_object.textLines(report_text)

    # Draw the text on the PDF
    c.drawText(text_object)

    # Save the PDF
    c.save()

# Example usage
patient_name = input("Enter the patient name")
age = int(input("enter patient age"))
gender = input("gender")
report_text = f"""{output}

Recommendations: Continue with current medication and follow up after one month.
"""

generate_hospital_report(patient_name, age, gender, report_text)
```
